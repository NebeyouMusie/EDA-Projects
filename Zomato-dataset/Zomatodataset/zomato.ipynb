```python
# let's see the other dataset called country code
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# let's check the columns of our df(zomato dataset)

df.columns
```

```python
# Now let's join the country code of the df(zomato dataset) and the df_country(country code dataset)

final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
# to check data types

final_df.dtypes
```

```python
country_names = final_df["Country"].value_counts().index
```

```python
country_names
```

```python
country_value = final_df["Country"].value_counts().values
```

```python
country_value
```

```python
# Let's visuallize the country names and their respective values using a pie chart

plt.pie(country_value, labels=country_names, autopct='%1.2f%%')
```

- because there are a number of countries, you can see that on the right side of the pie chart the texts(country names) for the respective values are cluttered, so inorder to give a clear visulization we can choose the top 3 countries that uses Zomato 

```python
plt.pie(country_value[:3], labels=country_names[:3], autopct="%1.2f%%")
```

### Observation

- From the above pie chart we can conclude that Zomato is highly used in India, United States then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
ratings
```

### Observation
- When rating is:
1. 4.5 - 4.9 it's Excellent and expressed by Dark Green color
2. 4.0 - 4.4 it's Very Good and expressed by Green color
3. 3.5 - 3.9 it's Good and expressed by Yellow color
4. 2.5 - 3.4 it's Average and expressed by Orange color
5. 1.8 - 2.4 it's Poor and expressed by Red color
6. 0.0 it's Not rated and expressed by White color

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating count", hue="Rating color", data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

- Inorder to make the value for the white rating visible I have changed it to a blue color

### Observation

1. The value for the Not Rated is very high (>2000)
2. Maximum number of rating is between 2.5 to 3.4

```python
# count plot

sns.countplot(x="Rating color", data=ratings,  palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
